#### pandas 라이브러리에서 외부의 데이터를 로드하는 함수들
-csv, tsv
    - 가장 단순한 형태의 데이터 파일
    - pandas에 내장된 read_csv() 함수를 이용해서 데이터 로드
    - csv는 데이터를 ',' 기준으로 나눠준다.
        - read_csv(경로)를 이용하여 로드
    - tsv는 데이터를 '\t'(탭) 기준으로 나눠준다.
        - read_csv() 함수에서 매개변수 sep 매개변수에 인자값을 '\t'로 설정
- json
    - 많이 사용이 되는 데이터 파일(특히 웹에서 많이 사용)
    - python으로 생각하면 dict 형태와 같다.
    - pandas에서 read_json() 함수를 이용하여 데이터 로드
- xml
    - html 구조 언어와 같이 tag를 이용해서 데이터가 구성
        - ex> <data>'abc'<\data>
    - pandas에서 read_xml()함수를 이용하여 데이터 로드
- 위의 3개의 read_xxx() 함수들은 공통적으로 encoding이라는 매개변수가 존재
    - 기존값은 'UTF-8'
    - 데이터 파일을 로드하는데 해당 데이터셋에 한글이 존재하는 경우에 종종 에러가 발생(decode error)
        - 인자값을 'CP949', 'EUC-KR'로 사용하면 데이터 로드가 가능
    - 엑셀 파일을 로드하는 함수에는 encoding 매개변수가 존재하지 않는다.
- excel(xls, xlsx)
    - 엑셀 데이터 파일
    - pandas 안에서 read_excel() 함수를 이용해서 데이터 로드
    - 특정 엑셀 파일을 로드 할때 추가적인 라이브러리 설치가 필요한 경우가 존재
        - 특정 라이브러리가 존재하지 않는다면 모듈을 찾을수 없습니다. 라는 에러가 발생
            - 에러 발생시 해당 라이브러리 설치 (openpyxl)
- 모든 read_xxx() 함수에서 첫번째 인자값은 파일의 경로를 입력

#### 결측치의 처리
- 결측치 데이터를 제거
    - 결측치가 포함되어있는 컬럼을 제거
        -해당 컬럼의 데이터가 중요하지 않은 경우
    - 결측치가 포함되어있는 인덱스를 제거
        - 결측치가 포함되어있는 인덱스의 수가 전체 데이터의 수보다 일반적으로 5% 미만인 경우
        - 결측치가 포함된 인덱스의 데이터 신뢰도가 떨어지는 경우
    - 결측치를 제거하는 함수
        -결측치(na) + 버리다/지우다(drop)
    - dropna()
        - 매개변수 axis : 기준이되는 축 지정 기본값은 rows (0)
            - 결측치가 포함되어있는 인덱스를 제거
            - 값을 columns나 1로 변경하면 결측치가 포함된 컬럼을 제거
        - 매개변수 inplace : 기본값은 False
            - 원본 데이터( class 내부에 저장된 변수 데이터 )를 변경할 것인가 지정
    - 결측치를 특정한 값으로 채워준다
        - 채우다(fill) + 결측치(na)
        - fillna()
            - 인자에 특정 데이터를 입력
                -결측치가 해당 인자 값으로 치환
            - 매개변수 method 사용
                - ffill : front fill -> 결측치 전의 데이터로 채워준다
                - bfill : back fill -> 결측치 후의 데이터로 채워준다

### 데이터 프레임의 필터
- 데이터프레임명.loc[ 인덱스의 조건식, 컬럼의 조건식 ]
    - 인덱스의 값을 기준으로 필터 (인덱스가 같은 경우에만(True) 출력)
    - 컬럼의 값을 기준으로 필터 (컬럼이 같은 경우에만(True) 출력)
    - 시작 : 종료 -> 시작값부터 종료값까지 (종료가 포함)
- 데이터프레임명.iloc[ 인덱스의 위치, 컬럼의 위치 ]
    - 인덱스의 위치를 기준으로 필터
    - 컬럼의 위치를 기준으로 필터
    - 시작 : 종료 -> 시작 위치부터 종료 위치 전까지 (종료 위치는 미포함)
- 특정 컬럼의 데이터를 확인
    - 데이터프레임[단일 컬럼] -> 조건이 단일이라면 데이터프레임은 스리즈로 변환하여 추출
    - 데이터프레임[ [컬럼1, 컬럼2, ...] ] -> 조건이 복수의 타입(리스트나 튜플)이라면 데이터프레임으로 추출

```python
# 결측치가 포함되어있는 데이터 필터링
df.loc[flag_null, ]
```

```python
# 결측치가 포함되어있지 않은 데이터를 확인
# not() -> 부정 함수 -> T을 F, F를 T -> 스리즈에서는 사용이 불가능
# not(flag_null)
# 스리즈에서 데이터들을 모두 부정처리 하는 방법 
# 음수처리하기 
# -flag_null
# 비트연산자(~) 사용 -> 2진수의 0과 1을 뒤집는다.
# TFT -> 101 (비트연산자를 사용) -> 010
~flag_null
```

```python
# 반복문을 이용하여 Series의 value들을 부정 
# Series 안에는 2개의 변수 확인 index, values
# values 데이터들을 하나씩 추출하여 not() 함수를 이용해서 부정 처리
flag_list = []
for i in flag_null:
    # print(not(i))
    flag_list.append(not(i))
flag_list
```

```python
df.loc[flag_list, ]
```

```python
# map() -> 1차원 데이터를 각각 원소별 변환이 필요할때 사용하는 함수
# map( 함수, 1차원 데이터 )

list_1 = [1,2,3,4]

def func_1(x):
    return x ** 2

list(
    map(
        func_1, list_1
    )
)
```

```python
# map() + lambda 함수 
# lambda 함수 : 이름 없는 함수(단순한 계산식을 사용할때)
# lambda 매개변수 : 매개변수의 연산
 
list(
    map(
        lambda x : x ** 2, 
        list_1
    )
) 
```

```python
# Serise 안에 map()함수가 존재 
flag_null.map(
    # x라는 매개변수에는 True, False, True 순차적으로 입력
    lambda x : not(x)
)
# lambda 함수는 연산이 간단한 경우에만 사용
```

```python
def func_2(x):
    return not(x)

flag_null.map(
    func_2
)
```
