```python
# 결측치를 채워준다 -> 어떻게 채울것인가?
# 데이터를 확인 -> 특정 상품의 item_price가 결측치가 아닌 데이터와 결측치인 데이터를 확인 
# 조건식1 : 특정 상품명( uriage['item_name'] == '상품A' ) , 
# 조건식2 : 결측치인가 ( uriage['item_price'].isna() )
flag = uriage['item_name'] == '상품A'
flag_null = uriage['item_price'].isna()
```

```python
# 상품명이 상품A(조건식1) 이고(그리고) 가격이 결측치가 아닌(조건식2의 부정) 데이터를 확인 
# bool 형태로 이루어진 Series의 부정은 : 음수처리(-), 비트연산자(~)
uriage.loc[
    flag & ~flag_null, 'item_price'
].value_counts()
```

```python
flag2 = uriage['item_name'] == '상품C'
```

```python
uriage.loc[
    flag2 & ~flag_null, 'item_price'
].value_counts()
```

```python
# 결측치에 데이터를 채운다
# 상품A 인 데이터들의 item_price의 평균 값을 
# 상품A 인 데이터에서 item_price가 결측치인 부분에 채워준다. 

# 상품명이 상품A이고 item_price가 결측치가 아닌 데이터에서 item_price의 평균
mean_price = uriage.loc[flag & ~flag_null, 'item_price'].mean()
mean_price
```

```python
# fillna() 사용 
# 상품명이 상품A이고 item_price가 결측치인 데이터에 결측치를 mean_price로 채워준다. 
# 스리즈 형태로 출력 -> 데이터가 변화X -> 대입이 필요
uriage.loc[flag & flag_null, 'item_price'] = \
    uriage.loc[flag & flag_null, 'item_price'].fillna(mean_price)
```

```python
# flag_null은 결측치를 채우기 전에 만든 결측치인 조건식
uriage.loc[flag & flag_null, 'item_price']
```

```python
mean_price2 = uriage.loc[
    flag2 & ~flag_null, 'item_price'
].mean()
# 상품C 이고 가격이 존재하는 데이터의 평균 (결측치를 제외하고 평균을 구하겠다)
# pandas에서 결측치를 포함한 연산은 결측치를 제외하고 연산
mean_price2
```

```python
# Series(상품C의 가격이 결측치인) = 단일 데이터(평균값)  --> Series의 모든 value가 단일데이터로 채워진다. 
uriage.loc[
    flag2 & flag_null, 'item_price'
] = mean_price2
```

```python
uriage.loc[
    flag2 & flag_null, 'item_price'
]
```

```python
# 모든 상품 이름을 기준으로 반복문 실행 
# 각 상품의 가격의 평균을 구하고 해당 값을 결측치에 채워준다 
# 모든 상품의 목록 필요(unique)
item_list =uriage['item_name'].unique()
item_list
```

```python
# 조건식 2 생성 (item_price가 결측치인) -> 상품명과 상관 없이 item_price의 결측치
flag_null = uriage['item_price'].isna()

for item_name in item_list:
    # item_list의 각 원소들이 대입 
    # print(item_name)
    # 조건식 1 생성 -> 반복실행할때마다 다른 조건식
    flag = uriage['item_name'] == item_name
    # 가격의 평균을 구한다. 
    mean_price = uriage.loc[
        flag & ~flag_null, 'item_price'
    ].mean()
    # flag와 flag_null을 모두 만족하는 item_price에 mean_price를 대입 
    uriage.loc[
        flag & flag_null, 'item_price'
    ] = mean_price
```

```python
# uriage에 item_price 결측치의 개수를 확인
uriage['item_price'].isna().sum()
```

```python
for i in range(len(item_list)):
    # item_list의 위치 값들이 대입
    # print(item_list[i])
    # 각각의 상품명에 따라 item_price의 평균 확인
    flag = uriage['item_name'] == item_list[i]

    mean_price = uriage.loc[flag, 'item_price'].mean()

    print(f"{item_list[i]} : {mean_price}")
```
