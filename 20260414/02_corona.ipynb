```python
new_list = []

# 819번 반복을 하는 반복문 
for i in range(819):
    # i는 0부터 818까지 대입
    result = corona.loc[i+1, '총 확진자'] - corona.loc[i, '총 확진자']
    # 계산 결과를 new_list에 추가 
    new_list.append(result)

new_list
```

```python
new_list2 = [0]

for i in range(1, 820, 1):
    result = corona.loc[i, '총 확진자'] - corona.loc[i-1, '총 확진자']
    new_list2.append(result)

len(new_list2)
```

```python
# 819번 반복하면서 일일확진자의 데이터는 819개 -> 첫번째 데이터는 일일확진자를 구할수 없기 때문
# 첫날의 데이터는 0으로 생성
# 만들어진 리스트에 0을 첫번째 데이터로 대입 -> list.insert(0, 0)
# 리스트 +  리스트 -> 연결 [0] + new_list
# 반복 실행하기 전 new_list를 빈 리스트가 아니라 [0] 대입한 뒤 append()
# case1
[0] + new_list
```

```python
# corona 데이터프레임에 new_list로 만들어준 일일확진자를 대입 
# 데이터프레임은 dict와 비슷한 특성을 가지고 있다. 
# 데이터프레임[new_col] = 1차원 데이터(Series, list, tuple) // 단일 데이터
# 1차원 데이터를 대입하는 경우는 데이터프레임의 인덱스와 길이가 같아야 한다. 
# 단일 데이터를 대입하는 경우 -> 모든 값이 단일데이터로 고정 
corona['일일 확진자'] = new_list2
```

```python
# 인덱스 이동(shift) 함수 
# shift()는 기본값은 인덱스는 1씩 이동 (아래로 한칸씩 이동)
# 1차원 데이터중 list - list ?? --> 에러 발생
# numpy의 ndarray형태(행렬) -> 행렬 - 행렬 -> 같은 위치의 원소끼리의 차를 구해서 행렬로 결과
# pandas가 설치될때 numpy가 같이 설치 --> pandas에서 numpy를 사용한다. 
# DataFrame과 Series의 values 값들은 행렬로 표시
# series - series --> 행렬의 차로 계산
# 54 - NaN = NaN // 결측치를 0으로 채운다. 
# 54 - 0 = 54
corona['일일 사망자'] = (corona['총 사망자'] - corona['총 사망자'].shift()).fillna(0)
# int - int -> int 
# 총사망자 shift()를 이용하여 계산을 하면 결측치가 우선 발생 (해당 데이터의 타입이 float 변환)
```

```python
corona.head()
```

```python
# 같은 컬럼내에서 데이터들간의 차이를 구하는 함수 
# diff(n) : n만큼 이동한 데이터와의 차이를 구하는 함수
# diff(1)의 결과는 Series - Series.shift(1) 결과와 같다
corona['일일 사망자2'] = corona['총 사망자'].diff(1)
```
