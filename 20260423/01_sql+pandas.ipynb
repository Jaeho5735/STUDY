```python
# emp, dept 테이블
# 테이블 간의 결합 -> 사원의 정보 테이블과 부서의 정보 테이블 -> 조건에 맞게 결합이 되는 형태 -> 조인 결합 
# 조인 결합 -> pandas 에서는 merge() 함수를 이용
```

- 유니언 결합 
    - 단순 결합 (행, 열)
    - pandas에서는 concat( [데이터프레임, 데이터프레임, ...] ) 함수를 이용
    - 단순 결합의 형태이기 때문에 데이터프레임의 개수의 제한 x
        - 매개변수 
            - axis 
                - 행으로 결합한 것인가, 열로 결합할 것인가
            - ignore_index 
                - 결합되는 축을 초기화 할것인가
    - SQL에서는 UNION 키워드를 이용해서 결합 
        - SELECT문 UNION SELECT문
- 조인 결합 
    - 특정한 조건에 맞춰서 열을 결합 
    - 데이터프레임과 데이터프레임간의 결합  -> 특정 컬럼의 데이터들이 같은 값들로 이우러져있는 경우 열을 추가하는 방식 
    - pandas, DataFrame에 내장된 merge() 함수를 이용
    - 조건에 맞춰서 결합이 되기 때문에 데이터프레임은 2개 결합 
        - 매개변수 
            - on
                - 조건식 -> 2개의 데이터프레임이 모두 가지고 있는 컬럼의 이름 
                - 각 데이터프레임의 컬럼이 이름이 다른 경우에는 left_on, right_on 매개변수를 이용
            - how
                - 조인 결합의 방식 (left, right, inner, outer)
    - SQL에서는 JOIN 키워드를 이용해서 결합 
        - SELECT * FROM `table1` INNER JOIN `table2` ON `table1`.`COL` = `table2`.`COL`

```python
# WHERE(인덱스 조건 - 필터) , GROUP BY(그룹화), HAVING(인덱스 조건 - 필터), ORDER BY(정렬)
# 1. WHERE(그룹화를 하기 전에 데이터를 필터)
# 2. GROUP BY(그룹화)
# 3. HAVING(그룹화 된 데이터에서 필터)
# 4. ORDER BY(최종 결과를 이용하여 정렬을 변경)
# 직종에서 SALESMAN을 제외하고 직종별로 그룹화를 하여 SAL의 평균을 구한뒤 평균의 값이 2500 이상인 데이터에서 평균 SAL이 높은 순서대로 출력

query_17 = """
    SELECT 
    `JOB`, AVG(`SAL`) `MEAN`
    FROM `emp`
    WHERE `JOB` != 'ANALYST' 
    GROUP BY `JOB`
    HAVING `MEAN` >= 2500 
    ORDER BY `MEAN` DESC
"""

sql(query_17)
```
