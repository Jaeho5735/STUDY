```python
tran_1 = pd.read_csv(r'C:\MULTICAM_11\csv\tran_1.csv')
tran_2 = pd.read_csv(r'C:\MULTICAM_11\csv\tran_2.csv')
```

```python
# tran_1, tran_2 데이터를 결합
    # 두 개의 데이터 프레임은 테이블의 구조가 같은 형태
    # 단순한 행 결합
    # 두 개의 데이터프레임이 행 결합을 하면 rows의 개수는? 5000 + 1786 = 6786
# concat() 함수 이용
# pandas 안에 concat 함수 호출 --> 원본의 데이터라는 개념이 존재 x -> 결과를 저장하려면 새로운 변수에 대입
tran = pd.concat(
    [tran_1, tran_2], axis = 0, ignore_index = True
)
```

```python
tran.info()
```

```python
tran_d_1 = pd.read_csv(r'C:\MULTICAM_11\csv\tran_d_1.csv')
tran_d_2 = pd.read_csv(r'C:\MULTICAM_11\csv\tran_d_2.csv')
```

```python
tran_d_1.info()
```

```python
tran_d_2.info()
```

```python
# tran_d_1, tran_d_2 데이터 프레임을 단순 행 결합
tran_d = pd.concat(
    [tran_d_1, tran_d_2], axis = 0, ignore_index = True
)
```

```python
tran_d.info()
```

```python
tran.head(1)
```

```python
tran_d.head(1)
```

```python
# tran과 tran_d 데이터 프레임을 결합
# 테이블의 구조는 다르게 구성 -> 단순한 행 결합은 불가능
# 인덱스의 개수가 다르다 -> 6786 // 7144개의 데이터
# 두 개의 데이터 프레임이 공통적으로 가지고 있는 데이터 컬럼은 존재하는가?
    # transaction_id 컬럼은 공통적으로 가지고 있다. -> 조인결합이 가능
# tran의 인덱스의 수는 6786개
# tran_d의 인덱스의 수는 7144개
    # 두 개의 데이터 프레임을 조인 결합 했을 때 => 인덱스의 수는 어떻게 나올것인가?
    # 6786 // 7144 --> how에 따라 다르게 나올것인가?
transaction = pd.merge(
    tran, tran_d,
    # 조건 (두개의 데이터 프레임이 공통적으로 가지고 있는 컬럼의 이름)
    on = 'transaction_id',
    # 결합의 방식은 어떻게 할 것인가?
    how = 'outer'
)
```

```python
test_1 = pd.merge(
    transaction, item_master,
    on = 'item_id',
    how = 'inner'
)
```

```python
customer_master = pd.read_csv(r'C:\MULTICAM_11\csv\customer_master.csv')
```

```python
customer_master.head(1)
```

```python
test_2 = pd.merge(
    test_1, customer_master,
    on = 'customer_id',
    how = 'inner'
)
```

```python
# 시계열 데이터 payment_date를 컬럼을 타입을 변경
# 문자를 시계열로 변경 (datetime.strptime() // to_datetime())
pd.to_datetime(test_2['payment_date'])
```

```python
# datetime 라이브러리의 strptime()을 이용하여 데이터 타입 변환
from datetime import datetime
```

```python
# datetime 안에 strptime() 함수는 pandas에 to_datetime()과는 다르게 스리즈 형태 변환 불가능
# datetime.strptime(test_1['payment_date'][0], '%Y-%m-%d %H:%M:%S')
# 스리즈 데이터에서 value들을 하나씩 뽑아서 함수에 넣는다.
test_1['payment_date'].map(
    # x에 대입이 되는 데이터는 문자
    lambda x : datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
)
```

```python
test_2['payment_date'] = pd.to_datetime(test_2['payment_date'])
```

```python
test_2.info()
```

```python
# 요일별 판매가 된 제품들의 총 판매 개수 합계가 어떻게 되는가?
# 요일이라는 새로운 변수를 생성 -> 시계열에서 요일만 따로 추출 (strftime(%A|%a|%w))
test_2['요일'] = test_2['payment_date'].dt.strftime('%A')
```

```python
test_2['payment_date'].dt.weekday
```

```python
# 요일, item_id를 기준으로 그룹화 -> quantity의 합산
test_2[['요일', 'item_id', 'quantity']].groupby(['요일', 'item_id']).sum()
```
