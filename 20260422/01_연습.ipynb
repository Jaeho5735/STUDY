```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get("http://moons-86.iptime.org:8080/")
```

```python
html_text = res.text()
soup = bs(html_text, 'html.parser')

div_data = soup.find_all(
    'div',
    attrs = {
        'id' : 'product-1001'
    }
)

div_data.get_text().split('\n')

# h2, p 태그륾 모두 찾아서 각각 텍스트 추출

p_list = div_data.find_all(
    ['h2', 'p']
)

[p.get_text() for p in p_list]

div_list = soup.find_all(
    'div',
    attrs = {
        'class' : 'product-item'
    }
)
```

```python
values = []

for div in div_list:
    p_list = div.find_all(
        ['h2', 'p']
    )
    value = [p.get_text() for p in p_list]
    values.append(value)
```

```python
df = pd.DataFrame(values, columns = ['상품명', '카테고리', '가격', '평점'])
```

```python
df.map(
    lambda x : x.split(':')[-1].lstrip()
)
```
