```python
# 조건문(if문)
# if 뒤의 조건식에 따라 참인 경우에 실행하는 코드와
# 거짓인 경우 실행하는 코드를 나눠서 작성
# python -> if 조건식 :
# 일반적인 언어 -> if (조건식) {
a = 5
# if문의 시작
if a < 10:
    # 조건식이 참인 경우 실행할 코드를 작성
    print('a는 10보다 작다')
# 거짓인 경우 실행할 코드
else:
    print('a는 10보다 크거나 같다')
# if ~ else 문에서는 하나의 지역의 코드만 실행이 된다.
```

```python
# 조건문에서 조건식이 여러개로 나눠지는 경우
score = 73 # C 학점

if score >= 90:
    # score가 90점 이상인 경우에만 실행
    print('A학점')
else:
    # score가 90점 미만인 경우 실행하는 지역
    if score >= 80:
        # score가 90점 미만이고 80점 이상인 경우 실행
        print('B학점')
    else:
        # score가 80점 미만인 경우
        print('C학점')
```

```python
# if ~ else를 사용 -> if 뒤의 조건식의 개수를 변경
if score >= 90:
    print('A학점')
# B학점의 조건식 : score가 90점 미만(조건식1) 그리고 80점 이상(조건식2)
if (score < 90) and (score >= 80):
    print('B학점')
# C학점의 조건식 : score가 80점 미만 그리고 70점 이상
if (score < 80)  and (score >= 70):
    print('C학점')
# F학점의 조건식 : score가 70점 미만
if score < 70:
    print('F학점')
```
