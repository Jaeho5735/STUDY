### 자료형 데이터
- 여러개의 데이터를 하나의 변수에 대입하기 위한 형태

- tuple
    - 자료형 데이터 중에 가장 기본적인 형태 ( 가장 메모리를 덜 먹는 데이터의 형태 )
    - ()를 이용해서 데이터를 묶여서 표현(생성)
    - 한번 생성이 되면 수정이 불가능 ( 원소를 수정하거나 추가 삭제 작업을 할때는 새로운 tuple를 생성 )
- list
    - []를 이용해서 데이터를 묶여서 표현(생성)
    - tuple과는 다르게 각 원소들을 수정하거나 추가, 삭제가능(리스트에서 제공하는 내장 함수를 이용)
    - 원소의 개수를 0개부터 생성이 가능 ( 비어있는 리스트를 생성 -> 여러 개의 데이터를 하나의 리스트에 추가하여 새로운 데이터의 묶음 생성 )
- dict
    - {}를 이용해서 데이터를 묶여서 표현(생성)
    - key : value의 형태로 데이터가 구성 ( list에서 index를 key 표현한 방식 )
    - key을 기준으로 데이터를 선택, 수정, 삭제 등이 가능 
    - 원소의 개수를 0개부터 생성이 가능

```python
def func_12(*data, types='max'):
    if types == 'max':
        return max(data)
    elif types == 'min':
        return min(data)
    elif types == 'mean':
        sum = 0
        for i in data:
            sum += i
        return sum // len(data)
    else:
        return print("types 데이터 입력값이 잘못되었습니다.")
```

```python
print(func_12(1,5,7))
print(func_12(1,4,7, types='min'))
print(func_12(1,4,7, types='mean'))
print(func_12(1,4,7, types='sub'))
```

```python
# 리스트형 하나 생성 
list_a = [10, 20, 30, 40]
list_b = ['a', 'b', 'c']
```

```python
# 리스트의 확장 (두개의 리스트를 결합)
# tuple에서 +를 이용하면 두개의 tuple이 결합 -> 리스트에서도 가능한가?
list_a + list_b
```

```python
# 리스트 내장함수 확장 extend() 함수가 존재
list_a.extend(list_b)
```

```python
# 다차원 리스트 생성 : index(위치(기준)가 여러개인 데이터)
list_c = [
    [10, 20, 30], 
    [40, 50, 60]
]   # 2차원 리스트
print(list_c)
```

```python
# list_c의 길이(원소의 개수)는 몇인가? -> 2
print(len(list_c))
```

```python
# 2차원 리스트 데이터에서 단일 데이터로 접근 -> 위치가 2개 -> 2번의 위치 선택 
# 20이라는 데이터에 접근하려면 -> index 0인 위치 안에 index 1인 위치에 데이터가 존재
print(list_c)       # 2차원 데이터
print(list_c[0])    # 1차원 데이터
print(list_c[0][1]) # 0차원 데이터(단일 데이터)
```
