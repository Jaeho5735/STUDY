#### 웹 크롤링
- 라이브러리
    - requests

    - bs4
        - BeautifulSoup class를 이용

            - 태그별 속성의 값
            - 태그의 특징
    - selenium
        - 웹 어플리케이션을 테스트하기 위한 라이브러리
        - 웹 브라우저(chrome, ...)를 python code를 이용하여 제어
            - 함수를 이용해서 특정 버튼을 클릭
            - 특정 위치에 데이터를 입력
        - 구글 크롬의 버전이 구버전인 경우 별도의 소프트웨어 설치 필요
- 데이터를 불법적으로 가지고 오는 방법 (합법 / 위법 사이)
    - 사이트들마다 크롤링이 가능한 주소들은 지정
    - 상업적인 용도가 아닌 수업에서의 내용은 큰 지장은 없음

#### 태그의 이름을 기준으로 검색
- soup.태그명 
    - 해당 html 문서 안에서 해당 태그 중 첫번째 태그를 되돌려준다.
    - 문자열 함수에서 find() 함수와 비슷한 부분.
- soup.태그명.string
    - 첫번째 태그에서 콘텐츠(태그 안에 데이터)를 되돌려준다.
        - ex) <p>Test</p> --> 'Test'
- soup.태그명[속성명]
    - 첫번째 태그에서 특정 속성의 값을 되돌려준다.
        - ex) <a href = "https://www.naver.com">네이버</a> -> https://www.naver.com

- 내장함수
    - find()
        - html 문서 안에서 특정 태그의 첫번째 정보를 출력
        - find(속성명 = 속성값)
            - 태그들 중 특성 속성에 특정한 값을 가진 태그의 첫번째 정보를 출력
        - 함수의 결과 값의 타입은 TAG 데이터의 타입으로 되돌려준다.
    - find_all()
        - html 문서 안에서 특정 태그의 모든 정보를 출력
        - limit 매개변수
            - 찾는 태그의 개수를 지정
        - 함수의 결과 값의 타입은 ResultSet 타입으로 되돌려준다.
    - find(), find_all() 함수는 Tag 타입에서는 사용이 가능, ResultSet에서는 사용이 불가능
    - ResultSet 타입은 Tag 데이터들이 모여있는 리스트의 형태와 흡사

```python
# html 문서 안에서 태그명이 div이고 class 속성의 값이 section_sise_top인 태그를 모두 찾아서 길이를 구하자
len(
    soup.find_all(
        'div',
        attrs = {
            'class' : 'section_sise_top'
        }
    )
)
```

```python
# 1개 라면 find_all()이 아닌 find() 이용해서 데이터 추출
div_data = soup.find(
    'div',
    attrs = {
        'class' : 'section_sise_top'
    }
)
```

```python
div_data
```

```python
# div_data에서 table 태그중 class의 값이 tbl_home인 태그의 개수가 어떻게 되는가?
len(
    div_data.find_all(
        'table',
        attrs = {
            'class' : 'tbl_home'
        }
    )
)
```

```python
# 8개의 테이블을 모두 찾는다.
tables = div_data.find_all(
    'table',
    attrs = {
        'class' : 'tbl_home'
    }
)
```

```python
tables[0]
# thead는 테이블의 제목 (columns 이름) -> 따로 추출하여 columns 사용
# tbody는 테이블의 데이터 (values) -> 따로 추출해서 values 사용
```

```python
# 반복문에서 사용할 변수명을 그대로 이용하여 1차 작업 (반복 실행 코드를 테스트)
table_data = tables[0]
```

```python
# columns에서 사용할 데이터들을 추출 -> table 데이터에서 thead 태그를 추출
head_data = table_data.find('thead')
```

```python
head_data
```

```python
# head_data에서 th 태그를 모두 찾는다.
th_list = head_data.find_all('th')
# 각각 th 태그에서 콘텐츠를 추출하여 리스트로 만든다.
# 콘텐츠를 추출하는 방법은 string // get_text()
th_list[0].get_text()
```

```python
# case 1
cols = []
for th in th_list:
    cols.append(th.string)
cols
```

```python
# case 2
cols2 = list(map(
    lambda x : x.string,
    th_list
))
cols2
```

```python
# case 3
cols3 = [ th.string for th in th_list ]
cols3
```

```python
# values에서 사용할 데이터들을 추출
body_data = table_data.find('tbody')
body_data
```

```python
# tr을 모두 찾는다.
tr_list = body_data.find_all(
    'tr'    
)
len(tr_list)
```

```python
# 첫번째 tr 데이터에서 value를 추출 -> 15번 반복
tr = tr_list[0]
tr
```

```python
# 두 개의 태그를 모두 찾는다?
row_list = tr.find_all(['th', 'td'])
```

```python
# row_list에서 콘텐츠 데이터를 추출 -> string, get_text()
# get_text() : 태그 안에 모든 콘텐츠를 추출
# string : 콘텐츠 데이터가 한군데에 존재할때 추출가능
row_list[2].get_text()
```

```python
value_data = [ data.get_text().strip() for data in row_list ]
value_data
```

```python
values = []

for tr in tr_list:
    row_list = tr.find_all(['th', 'td'])
    value_data = [ data.get_text().strip() for data in row_list ]
    # value_data를 이용해서 2차원 데이터 values 생성
    values.append(value_data)
values
```

```python
# 8개의 테이블 중 1개 완성
# 8번 반복을 하는 반복문을 생성 -> 각각의 데이터 프레임이 생성 -> 각각 다른 전역 변수에 저장 (globals()["변수명"])

# 전역변수의 이름을 반복 실행할때 마다 변경하기 위한 기본값
vari_num = 1

for table_data in tables:
    # thead를 추출 -> columns 데이터를 추출하기 위해서
    head_data = table_data.find('thead')
    # thead 데이터에서 각각 콘텐츠 데이터를 추출하여 리스트화
    cols = [ th.string for th in th_list ]

    # tbody를 추출
    body_data = table_data.find('tbody')
    # 여러개의 행이 존재 -> tr_list를 생성
    tr_list = body_data.find_all('tr')
    # 각각의 tr_data를 추출하여 콘텐츠 데이터를 리스트화 하고 새로운 리스트에 추가
    values = []

    for tr in tr_list:
        row_list = tr.find_all(['th', 'td'])
        value_data = [ data.get_text().strip() for data in row_list ]
        # value_data를 이용해서 2차원 데이터 values 생성
        values.append(value_data)
    
    # 데이터 프레임을 생성하고 전역변수에 저장
    globals()[f"df{vari_num}"] = pd.DataFrame(values, columns = cols)
    
    # vari_num을 1 증가시킨다.
    vari_num +=1
```
